```python
import pandas as pd
data = pd.read_excel('Capstone_final_dataset.xlsx')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
# Bar chart for first_tx (Treatment Type) by SEX (Gender)
treatment_by_gender = data.groupby(['first_tx', 'SEX']).size().unstack()

# Plotting the bar chart
treatment_by_gender.plot(kind='bar', stacked=True)
plt.title('Distribution of Treatment Type by Gender')
plt.xlabel('Treatment Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for age category (age_cat)
age_cat_counts = data['age_cat'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(6,6))
plt.pie(age_cat_counts, labels=age_cat_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Age Category Distribution')
plt.show()
```

```python
# Scatter plot for all features (age, ABIRATERONE, enza)
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['ABIRATERONE'], label='Abiraterone', alpha=0.6)
plt.scatter(data['age'], data['enza'], label='Enzalutamide', alpha=0.6)
plt.title('Scatter Plot: Age vs Abiraterone/Enzalutamide Treatments')
plt.xlabel('Age')
plt.ylabel('Treatment')
plt.legend()
plt.show()
```

```python
# Box plot and histogram for two features: age and Total_Elixhauser_Groups
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Box plot for age
axes[0, 0].boxplot(data['age'].dropna())
axes[0, 0].set_title('Box Plot of Age')
axes[0, 0].set_ylabel('Age')

# Box plot for Total Elixhauser Groups
axes[0, 1].boxplot(data['Total_Elixhauser_Groups'].dropna())
axes[0, 1].set_title('Box Plot of Total Elixhauser Groups')
axes[0, 1].set_ylabel('Total Elixhauser Groups')

# Histogram for age
axes[1, 0].hist(data['age'].dropna(), bins=10, color='skyblue', edgecolor='black')
axes[1, 0].set_title('Histogram of Age')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Frequency')

# Histogram for Total Elixhauser Groups
axes[1, 1].hist(data['Total_Elixhauser_Groups'].dropna(), bins=10, color='orange', edgecolor='black')
axes[1, 1].set_title('Histogram of Total Elixhauser Groups')
axes[1, 1].set_xlabel('Total Elixhauser Groups')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
